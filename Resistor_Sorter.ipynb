```python
import cv2 as cv
import sys
from google.colab.patches import cv2_imshow
```

```python
img= cv.imread('/content/starynight.jpg')
cv2_imshow(img)
```

```python
grey = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
cv2_imshow(grey)
```

```python
sobelx = cv.Sobel(img, cv.CV_32F, 1, 0)
cv2_imshow(sobelx)
```
